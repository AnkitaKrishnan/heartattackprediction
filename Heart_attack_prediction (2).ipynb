```python
heartattack_df = heartattack_df.drop(columns='slope')
heartattack_df = heartattack_df.drop(columns="ca")
heartattack_df = heartattack_df.drop(columns="thal")
```

```python
heartattack_df.head()
```

```python
def male_female(pateints):
    sex=pateints
    if sex == 1:
        return "Male"
    else:
        return "Female"
```

```python
heartattack_df[0:11]
```

```python
heartattack_df['Gender'] = heartattack_df['sex'].apply(male_female)
```

```python
def chestpain_type(value):
    cp = value
    if cp == 1:
        return "Typical Angina"
    elif cp == 2:
        return "Atypical Angina"
    elif cp == 3:
        return "Non-anginal Pain"
    else:
        return "Asymptomatic"
```

```python
heartattack_df['Chest_Pain_Type'] = heartattack_df['cp'].apply(chestpain_type)
```

```python
heartattack_df.head()
```

```python
sns.catplot('cp',data=heartattack_df,hue='Gender',kind="count")
```

```python
def diabetic(Diabetes):
    fbs = Diabetes
    if fbs == 1:
        return "Diabetic"
    else:
        return "Non-Diabetic"
```

```python
heartattack_df['Diabetes']=heartattack_df['fbs'].apply(diabetic)
```

```python
def cardio_elec(cardio):
    cg = cardio
    if cg == 0:
        return 'Normal'
    elif cg == 1:
        return 'Having ST/T wave abnormality'
    else:
        return "Definite left ventricular hypertrophy by Estes's criteria"
```

```python
heartattack_df['cardio'] = heartattack_df['restecg'].apply(cardio_elec)
```

```python
heartattack_df.head()
```

```python


h = sns.lmplot(
    x="age", 
    y="chol", 
    hue="cardio", 
    data=heartattack_df,  
    legend=False,
    aspect=3
)

plt.legend(title='''Resting Blood Pressure categorised to see differences
           in gender''', loc='upper left', labels=['Normal', 'Having ST/T wave abnormality',
                                                   "Definite left ventricular hypertrophy by Estes's criteria"])
plt.show(h)
```
